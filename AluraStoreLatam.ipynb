### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

tienda = pd.read_csv("base-de-datos-challenge1-latam/tienda_1.csv")
tienda2 = pd.read_csv("base-de-datos-challenge1-latam/tienda_2.csv")
tienda3 = pd.read_csv("base-de-datos-challenge1-latam/tienda_3.csv")
tienda4 = pd.read_csv("base-de-datos-challenge1-latam/tienda_4.csv")

tienda.info()
```

## 1. Análisis de facturación



### Tienda 1

```python
# Ventas Tienda 1
print("Ventas:",tienda["Precio"].sum())
print("Costo Envio:",tienda["Costo de envío"].sum())
print("Ganancia Neta:",tienda["Precio"].sum() - tienda["Costo de envío"].sum())
```

### Tienda 2

```python
# Ventas Tienda 2
print("Ventas:",tienda2["Precio"].sum())
print("Costo Envio:",tienda2["Costo de envío"].sum())
print("Ganancia Neta:",tienda2["Precio"].sum() - tienda2["Costo de envío"].sum())
```

### Tienda 3

```python
# Ventas Tienda 3
print("Ventas:",tienda3["Precio"].sum())
print("Costo Envio:",tienda3["Costo de envío"].sum())
print("Ganancia Neta:",tienda3["Precio"].sum() - tienda3["Costo de envío"].sum())
```

### Tienda 4

```python
# Ventas Tienda 4
print("Ventas:",tienda4["Precio"].sum())
print("Costo Envio:",tienda4["Costo de envío"].sum())
print("Ganancia Neta:",tienda4["Precio"].sum() - tienda4["Costo de envío"].sum())
```

### Grafica de Barras de Facturacion

```python
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
sns.barplot(x=["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], y=[tienda["Precio"].sum(), tienda2["Precio"].sum(), tienda3["Precio"].sum(), tienda4["Precio"].sum()])
plt.title("Ventas Totales por Tienda")
plt.ylabel("Ventas Totales")
plt.ylim(0.8,1180000000)
plt.xlabel("Tiendas")  
```

# 2. Ventas por categoría

### Ventas Tienda 1

```python
ventas_por_categoria1 = tienda.groupby("Categoría del Producto")["Precio"].sum().reset_index()
ventas_por_categoria1 = ventas_por_categoria1.sort_values("Precio", ascending=False)


plt.figure(figsize=(10,6))
sns.barplot(x=ventas_por_categoria1["Categoría del Producto"], y=ventas_por_categoria1["Precio"])
plt.title("Ventas por Categoría en Tienda 1")
plt.ylabel("Ventas")
plt.xlabel("Categorías")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
plt.figure(figsize=(10,6))
sns.barplot(x=["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], y=[tienda["Calificación"].mean(), tienda2["Calificación"].mean(), tienda3["Calificación"].mean(), tienda4["Calificación"].mean()])
plt.title("Calificaciones Totales por Tienda")
plt.ylabel("Calificaciones Totales")
plt.ylim(3.4,4.2)
plt.xlabel("Tiendas")
plt.tight_layout()
plt.show()
print("Promedio Tienda 1:", tienda["Calificación"].mean().round(3))
print("Promedio Tienda 2:", tienda2["Calificación"].mean().round(3))
print("Promedio Tienda 3:", tienda3["Calificación"].mean().round(3))
print("Promedio Tienda 4:", tienda4["Calificación"].mean().round(3))
```
